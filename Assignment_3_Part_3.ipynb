### 1. First thing to do is generating all correlation for all the pairs

### 2. Define a function that takes the whole cor-table as arguement and return a list of tuples with three elements in side each tuple. They are 1. exact correlation that is higher equal to 0.5, or lower equal to -0.5; and 2. the two names of the  pairs.

```python
def higherCor(correlation_table):
    var_list=list(correlation_table.columns)
    list_cor=[]
    for var1 in var_list:
        for var2 in var_list:
            cor_value=corr_table.loc[var1,var2]
            if cor_value>=0.5 or cor_value<=-0.5:
                list_cor.append((var1,var2,cor_value))
    return list_cor
```

```python
original_list=higherCor(corr_table)
```

```python
original_list### Note that this list contains all pairs within the cor-table including those that pairs up with it self E.g. ('Id','Id','1.0'), which are meaningless for the study.
```

### 3.Creating a new list with all unique pairs and corelation that is not equal to 1 based on the original list generated from the last step 

```python
def newlistcor(somelist):
    '''Creating a new list called new_list1 to contain those pairs that are not itself and have the correlation of 1''' 
    new_list1=[]
    '''Using for loop to iterate each tuple within the original list and only select those that is not pairing itself'''
    for pair in somelist:
        if pair[2]!=1.0:
            new_list1.append(pair)
    '''After having a list contained with pairs that formed with differet variables, we sort the list in ascending order according to the coefficient so that the duplicate pairs will stick together'''        
    new_list2=sorted(new_list1,key=lambda x: x[2])
    '''Once all the duplicate tuples have stick together, we now can use the sllicing method to only pick up the second item every two item to create a list that only contain the unique pair with its coefficient value'''
    final_list=new_list2[::2]
    return final_list
```

```python
strong_cor_list=newlistcor(original_list)
strong_cor_list
```
